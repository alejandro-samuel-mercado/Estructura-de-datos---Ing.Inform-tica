```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import re
import random
import time


def pedir_entrada(prompt):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el cuadro de diálogo de entrada
    entrada = simpledialog.askstring(title="Entrada", prompt=prompt)

    # Cerrar la ventana principal
    root.destroy()

    return entrada


def mostrar_error(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de error
    messagebox.showerror("Error", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def mostrar_info(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de información
    messagebox.showinfo("Información", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def validarTamanio():
    while True:
        try:
            tamanio = int(pedir_entrada(
                f"Ingrese el tamaño de la lista: "))
            if 0 <= tamanio <= 10:
                break
            else:
                mostrar_error("La nota debe estar entre 0 y 10")
        except ValueError:
            mostrar_error("Ingrese un valor correcto")
    return tamanio


def generarNumeros(lista):
    tamanio = validarTamanio()
    for i in range(tamanio):
        num = random.randint(1, 100)
        lista.append(num)


def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        mergeSort(left_half)
        mergeSort(right_half)

        i = j = k = 0

        # Mezcla las dos mitades
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Copia los elementos restantes de la mitad izquierda
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # Copia los elementos restantes de la mitad derecha
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr


def quickSort(lista, start, stop):
    # def quicksort(arr):
    # if len(arr) <= 1:
    #     return arr
    # else:
    #     pivot = arr[0]
    #     less_than_pivot = [x for x in arr[1:] if x <= pivot]
    #     greater_than_pivot = [x for x in arr[1:] if x > pivot]
    #     return quicksort(less_than_pivot) + [pivot] + quicksort(greater_than_pivot)

    if start < stop:
        # Elección del pivote
        pivot = lista[stop]
        i = start - 1

        # Reordenar los elementos alrededor del pivote
        for j in range(start, stop):
            if lista[j] <= pivot:
                i += 1
                lista[i], lista[j] = lista[j], lista[i]

        # Colocar el pivote en la posición correcta
        lista[i + 1], lista[stop] = lista[stop], lista[i + 1]
        pivot_index = i + 1

        # Recursión sobre las dos mitades
        quickSort(lista, start, pivot_index - 1)
        quickSort(lista, pivot_index + 1, stop)
    return lista


def shell(lista):
    n = len(lista)
    gap = n//2

    while gap > 0:
        for i in range(gap, n):
            tmp = lista[i]
            j = i
            while j >= gap and lista[j-gap] > tmp:
                lista[j] = lista[j-gap]
                j -= gap
            lista[j] = tmp
        gap = gap//2
    return lista


def mostrarLista(lista, metodo):
    datos = f"Lista ordenada por {metodo}\n"
    for e in lista:
        datos += f"{e} "

    mostrar_info(datos)


def main():
    lista = []
    generarNumeros(lista)
    mostrarLista(lista, "NADA")

  # Ordenamiento con Shell Sort
    lista_shell = lista.copy()
    start_time = time.time()
    ordenarPorShell = shell(lista_shell)
    merge_sort_time = round(time.time() - start_time, 6)
    mostrarLista(ordenarPorShell, "Shell")
    mostrar_info(f"Tiempo de ejecución {merge_sort_time}")

    # Ordenamiento con Quick Sort
    lista_quick = lista.copy()
    start_time = time.time()
    ordenarPorQuickSort = quickSort(lista_quick, 0, len(lista_quick) - 1)
    merge_sort_time = round(time.time() - start_time, 6)
    mostrarLista(ordenarPorQuickSort, "QuickSort")
    mostrar_info(f"Tiempo de ejecución {merge_sort_time}")

   # Ordenamiento con Merge Sort
    lista_merge = lista.copy()
    start_time = time.time()
    ordenarPorMerge = mergeSort(lista_merge)
    merge_sort_time = round(time.time() - start_time, 6)
    mostrarLista(ordenarPorMerge, "Merge Sort")
    mostrar_info(f"Tiempo de ejecución {merge_sort_time}")


main()
```

```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import re
import random
import time


def pedir_entrada(prompt):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el cuadro de diálogo de entrada
    entrada = simpledialog.askstring(title="Entrada", prompt=prompt)

    # Cerrar la ventana principal
    root.destroy()

    return entrada


def mostrar_error(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de error
    messagebox.showerror("Error", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def mostrar_info(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de información
    messagebox.showinfo("Información", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def validarNombre(mensaje):
    while True:
        nombre = pedir_entrada(mensaje)
        if nombre and re.match(r'^[a-zA-Z\s]+$', nombre):
            break
        else:
            mostrar_error("El nombre solo debe contener letras")
    return nombre


def validarTipo(mensaje):
    while True:
        nombre = pedir_entrada(mensaje).lower()
        if nombre and re.match(r'^[a-zA-Z\s]+$', nombre):
            if nombre == "herramienta" or nombre == "maquinaria":
                break
            else:
                mostrar_info("Ingrese un tipo válido")
        else:
            mostrar_error("El nombre solo debe contener letras")
    return nombre


def validarPrecio():
    while True:
        try:
            peso = float(pedir_entrada("Precio de alquiler por día "))
            break
        except (ValueError):
            mostrar_error("Valor no válido")
    return peso


def validarDisponibilidad():
    while True:
        disponible = pedir_entrada("Disponibilidad: True o False").lower()
        if disponible == "true" or disponible == "false":
            break
        else:
            mostrar_info("Ingrese un tipo válido")

        mostrar_error("El nombre solo debe contener letras")
    return disponible


def validarPorcentaje():
    while True:
        try:
            porcentaje = float(pedir_entrada(
                "Porcentaje a aumentar [0.1, 0.9]"))
            if 0.1 <= porcentaje <= 0.9:
                break
            else:
                mostrar_info("Ingrese un porcentaje válido")
        except:
            mostrar_error("Debe contener solo números")
    return porcentaje


def validarDias():
    while True:
        try:
            tamanio = int(pedir_entrada(
                f"Ingrese los días a alquilar: 1-200 "))
            if 0 <= tamanio <= 200:
                break
            else:
                mostrar_error("Ingrese un día válido")
        except ValueError:
            mostrar_error("Ingrese un valor correcto")
    return tamanio


def registrarMaquinaria(lista):
    nombre = validarNombre("Nombre")
    tipo = validarTipo("Tipo: herramienta/maquinaria")
    precio = validarPrecio()
    disponibilidad = validarDisponibilidad()
    lista.append([nombre, tipo, precio, disponibilidad])


def mostrarListado(lista, mensaje):
    dato = f""" {mensaje}
    -----------------------------------------------"""
    for elemento in lista:
        dato += f"""
    Nombre:{elemento[0]}
    Tipo: {elemento[1]}
    Precio por día: {elemento[2]}
    Disponibilidad:{elemento[3]}
    ------------------------------"""

    mostrar_info(dato)


def ordenarListadoPorPrecio(lista):
    listaHerramientas = [h for h in lista if h[1] == "herramienta"]
    n = len(listaHerramientas)
    gap = n//2

    while gap > 0:
        for i in range(gap, n):
            tmp = listaHerramientas[i]
            j = i
            while j >= gap and listaHerramientas[j-gap] > tmp:
                listaHerramientas[j] = listaHerramientas[j-gap]
                j -= gap
            listaHerramientas[j] = tmp
        gap = gap//2

    lista[:] = listaHerramientas

    mostrarListado(lista, "Herramientas ordenadas por precio")


def calcularPresupuestoAlquiler(lista):
    nombre = validarNombre("Nombre de la Herramienta/Maquinaria a alquilar")
    presupuesto = 0
    encontrado = False
    for elemento in lista:
        if elemento[0] == nombre:
            if elemento[3] == "false":
                mostrar_info(f"La {elemento[1]} {nombre} no está disponible")
                encontrado = True
                break
            elif elemento[3] == "true":
                cantidadDias = validarDias()
                presupuesto = cantidadDias*elemento[2]
                mostrar_info(f"""Presupuesto total:\n {presupuesto} """)
                encontrado = True
                break

    if not encontrado:
        mostrar_info("No se encontró la Herramienta/Maquinaria")


def mostrarTotalPorDisponibilidad(lista, disponibilidad):
    if disponibilidad == "true":
        mensaje = "Disponible"
    elif disponibilidad == "false":
        mensaje = "No disponible"

    dato = f"""Precio total de herramientas/maquinaria {mensaje}
    -----------------------------------------------\n"""
    total = 0

    for elemento in lista:
        if elemento[3] == disponibilidad:
            total += elemento[2]

    dato += f"{total}"
    mostrar_info(dato)


def actualizacionPrecios(lista):
    porcentaje = validarPorcentaje()

    for elemento in lista:
        elemento[2] += elemento[2]*porcentaje

    mostrarTotalPorDisponibilidad(lista, "true")
    mostrarTotalPorDisponibilidad(lista, "false")


def main():
    lista = []
    while True:
        try:
            opcion = int(pedir_entrada("""Seleccione una opción: 
                                     1: Registrar herramienta/maquina
                                     2: Mostrar listado de herramientas/maquinaria
                                     3: Ordenar el listado de herramientas/maquinas
                                     4: Realizar el presupuesto del alquiler 
                                     5: Realizar actualización de precios
                                     6: Salir 
                                     """
                                       ))
            if opcion:
                if opcion == 1:
                    registrarMaquinaria(lista)
                elif opcion == 2:
                    if len(lista) != 0:
                        mostrarListado(
                            lista, "Listado de Herramientas/Maquinaria")
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 3:
                    if len(lista) != 0:
                        ordenarListadoPorPrecio(lista)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 4:
                    if len(lista) != 0:
                        calcularPresupuestoAlquiler(lista)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 5:
                    if len(lista) != 0:
                        actualizacionPrecios(lista)
                elif opcion == 6:
                    mostrar_info("Saliendo.....")
                    break
            else:
                mostrar_info("Opción incorrecta")
        except ValueError:
            mostrar_error("Ingrese un número válido")
        pass


main()
```

```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import re


def pedir_entrada(prompt):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el cuadro de diálogo de entrada
    entrada = simpledialog.askstring(title="Entrada", prompt=prompt)

    # Cerrar la ventana principal
    root.destroy()

    return entrada


def mostrar_error(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de error
    messagebox.showerror("Error", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def mostrar_info(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de información
    messagebox.showinfo("Información", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def validar_entrada(entrada):
    if entrada.isalpha():
        mostrar_info(f"Entrada válida: {entrada}")
    else:
        mostrar_error("La entrada debe contener solo letras.")


def validarDni():
    while True:

        dni = pedir_entrada("Ingrese el DNI: ")
        if len(dni) == 8 and dni.isdigit():
            break
        else:
            mostrar_error(
                "El DNI debe tener hasta 8 dígitos y solo contener números.")

    return dni


def validarNombre():
    while True:
        nombre = pedir_entrada("Ingrese el nombre: ")
        if nombre and re.match(r'^[a-zA-Z\s]+$', nombre):
            break
        else:
            mostrar_error("El nombre solo debe contener letras")
    return nombre


def validarNota():
    while True:
        try:
            nota = int(pedir_entrada("Ingrese la nota: "))
            if 0 <= nota <= 10:
                break
            else:
                mostrar_error("La nota debe estar entre 0 y 10")
        except ValueError:
            mostrar_error("Ingrese un valor correcto")
    return nota


def cargarNotas(lista):
    dni = validarDni()
    nombre = validarNombre()
    nota = validarNota()

    alumno = {
        "DNI": dni,
        "Nombre y Apellido": nombre,
        "Nota": nota
    }

    lista.append(alumno)


def promedio(lista):
    suma = 0
    for i in range(len(lista)):
        elemento = lista[i]["Nota"]
        suma += elemento
    promedio = suma/len(lista)
    return promedio


def mostrarLista(lista):
    prom = promedio(lista)
    indice = 1
    for alumno in lista:
        mostrar_info(f"""
                     {indice} :
                     Nombre: {alumno["Nombre y Apellido"]}
                     Nota: {alumno["Nota"]}
        -------------------------------------------------
        """)
        indice+1

    mostrar_info(f"Promedio de notas:  {prom}")


def mostrarAlumnosMayorAlPromedio(lista):
    prom = promedio(lista)
    listaMayorAlPromedio = []

    for alumno in lista:
        if alumno["Nota"] >= prom:
            listaMayorAlPromedio.append(alumno)
    cantNotasMayorAlPromedio = len(listaMayorAlPromedio)

    print("Alumnos con notas mayor al promedio: ")
    for alumno in listaMayorAlPromedio:
        mostrar_info(f"""
                     1 :
                     Nombre: {alumno["Nombre y Apellido"]}
                     Nota: {alumno["Nota"]}
        -------------------------------------------------
        """)
    mostrar_info(f"Cantidad de notas mayores al promedio: {
                 cantNotasMayorAlPromedio}")


def mostrarAlumnosConNotaMayor(lista):
    while True:
        try:
            notaAElegir = int(pedir_entrada(
                "Ingrese una nota para mostrar a los alumnos con nota igual o mayor: "))
            if 0 <= notaAElegir <= 10:
                break
            else:
                mostrar_info("Nota incorrecta. Debe estar entre 0 y 10")

        except ValueError:
            mostrar_error("Ingrese solo números")
    listaMayorA = []
    for alumno in lista:
        if alumno["Nota"] >= notaAElegir:
            listaMayorA.append(alumno)

    if listaMayorA != 0:
        for alumno in listaMayorA:
            mostrar_info(f"""
                1 :
                Nombre: {alumno["Nombre y Apellido"]}
                Nota: {alumno["Nota"]}
   -------------------------------------------------
   """)
    else:
        mostrar_info("Ningún alumno sacó igual o mas que la nota ingresada ")


def calcularNotaMaxima(lista):
    notaMaxima = lista[0]["Nota"]
    posicionNotaMax = 0
    for i in range(1, len(lista)):
        elemento = lista[i]["Nota"]
        if elemento > notaMaxima:
            notaMaxima = elemento
            posicionNotaMax = i

    mostrar_info(f"""Mayor nota: {notaMaxima}
    Alumno con dicha nota: {lista[posicionNotaMax]["Nombre y Apellido"]}
                 """)


def main():
    listaAlumnos = []
    while True:
        try:
            opcion = int(pedir_entrada("""Seleccione una opción: 
                                     1: Cargar alumno
                                     2: Mostrar lista de alumnos
                                     3: Mostrar alumnos con nota mayor al promedio
                                     4: Mostrar alumnos con nota mayor a .....
                                     5: Nota máxima y alumno con dicha nota
                                     6: Salir 
                                     """
                                       ))
        except ValueError:
            mostrar_error("Ingrese un número válido")
        if opcion:
            if opcion == 1:
                cargarNotas(listaAlumnos)
            elif opcion == 2:
                if len(listaAlumnos) != 0:
                    mostrarLista(listaAlumnos)
                else:
                    mostrar_error("La lista está vacía")

            elif opcion == 3:
                if len(listaAlumnos) != 0:
                    mostrarAlumnosMayorAlPromedio(listaAlumnos)
                else:
                    mostrar_error("La lista está vacía")

            elif opcion == 4:
                if len(listaAlumnos) != 0:
                    mostrarAlumnosConNotaMayor(listaAlumnos)
                else:
                    mostrar_error("La lista está vacía")

            elif opcion == 5:
                if len(listaAlumnos) != 0:
                    calcularNotaMaxima(listaAlumnos)
            elif opcion == 6:
                mostrar_info("Saliendo.....")
                break
        else:
            mostrar_info("Opción incorrecta")


main()
```

```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import re


def pedir_entrada(prompt):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el cuadro de diálogo de entrada
    entrada = simpledialog.askstring(title="Entrada", prompt=prompt)

    # Cerrar la ventana principal
    root.destroy()

    return entrada


def mostrar_error(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de error
    messagebox.showerror("Error", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def mostrar_info(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de información
    messagebox.showinfo("Información", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def validar_entrada(entrada):
    if entrada.isalpha():
        mostrar_info(f"Entrada válida: {entrada}")
    else:
        mostrar_error("La entrada debe contener solo letras.")


def validarEdad():
    while True:
        try:
            edad = int(pedir_entrada("Edad: "))
            if 0 <= edad <= 20:
                break
            else:
                mostrar_error(
                    "LA edad debe estar entre o y 20, y solo contener números.")
        except (ValueError):
            mostrar_error("No es un número")
    return edad


def validarPeso():
    while True:
        try:
            peso = float(pedir_entrada("Peso "))
            if 0 <= peso <= 100:
                break
            else:
                mostrar_error("Peso no válido")
        except (ValueError):
            mostrar_error("No es un número")
    return peso


def validarTexto(mensaje):
    while True:
        nombre = pedir_entrada(mensaje)
        if nombre and re.match(r'^[a-zA-Z\s]+$', nombre):
            if 3 <= len(nombre) <= 30:
                break
            else:
                mostrar_error("El nombre debe tener de 3 a 30 carácteres")
        else:
            mostrar_error("Solo debe contener letras")
    return nombre


def validarTelefono():
    while True:
        try:
            telefono = pedir_entrada("Nº de teléfono: ")
            if 0 <= len(telefono) <= 10 and telefono.isdigit():
                break
            else:
                mostrar_error("El rango debe ser de hasta 10 números")
        except (ValueError):
            mostrar_error("No es un número")
    return telefono


def validarDni():
    while True:

        dni = pedir_entrada("Ingrese el DNI: ")
        if len(dni) == 8 and dni.isdigit():
            break
        else:
            mostrar_error(
                "El DNI debe tener hasta 8 dígitos y solo contener números.")

    return dni


def cargarDueño():
    nombre = validarTexto("""   Datos del dueño \n
    Nombre
                          """)
    dni = validarDni()
    telefono = validarTelefono()
    direccion = pedir_entrada("Dirección")
    dueño = {
        "Nombre": nombre,
        "Dni": dni,
        "Telefono": telefono,
        "Direccion": direccion
    }

    return dueño


def cargarMascotas(lista):
    nombre = validarTexto("Nombre de la mascota")
    especie = validarTexto("Especie")
    edad = validarEdad()
    peso = validarPeso()
    salud = validarTexto("Estado de salud")
    dueño = cargarDueño(lista)

    mascota = {
        "Nombre": nombre,
        "Especie": especie,
        "Edad": edad,
        "Peso": peso,
        "Salud": salud,
        "Dueño": dueño
    }

    lista.append(mascota)


def mostrarMascotas(lista):
    indice = 1
    for mascota in lista:
        mostrar_info(f"""
                     Mascota{indice} :
                     Nombre: {mascota["Nombre"]}
                     Especie: {mascota["Especie"]}
                     Edad: {mascota["Edad"]}
                     Peso: {mascota["Peso"]}
                     Salud:{mascota["Salud"]}
                     ------------------------
                     Datos del dueño

                     Nombre:{mascota["Dueño"]["Nombre"]}
                     Dni:{mascota["Dueño"]["Dni"]}
                     Telefono:{mascota["Dueño"]["Telefono"]}
                     Dirección:{mascota["Dueño"]["Direccion"]}
        -------------------------------------------------
        """)
        indice += 1


def buscarMascotaPorNombre(lista):
    nombreABuscar = pedir_entrada("Nombre de mascota a buscar")
    encontrado = False
    for mascota in lista:
        if mascota["Nombre"] == nombreABuscar:
            mostrar_info(f"""
                     Nombre: {mascota["Nombre"]}
                     Especie: {mascota["Especie"]}
                     Edad: {mascota["Edad"]}
                     Peso: {mascota["Peso"]}
                     Salud:{mascota["Salud"]}
                     ------------------------
                     Datos del dueño

                     Nombre:{mascota["Dueño"]["Nombre"]}
                     Dni:{mascota["Dueño"]["Dni"]}
                     Telefono:{mascota["Dueño"]["Telefono"]}
                     Dirección:{mascota["Dueño"]["Direccion"]}
        -------------------------------------------------
        """)
            encontrado = True
            break
    if encontrado == False:
        mostrar_info("No se encontró a la mascota")


def mostrarMascotasDeTalDueño(lista):
    dueño = pedir_entrada("Ingrese el nombre del dueño de la/las mascota/s ")
    encontrado = False
    for mascota in lista:
        if dueño == mascota["Dueño"]["Nombre"]:
            mostrar_info(f"""
                     Nombre: {mascota["Nombre"]}
                     Especie: {mascota["Especie"]}
                     Edad: {mascota["Edad"]}
                     Peso: {mascota["Peso"]}
                     Salud:{mascota["Salud"]}
        ------------------------------------
        """)
            encontrado = True
    if encontrado == False:
        mostrar_info("No se encontró al dueño")


def mostrarMascotasEnfermas(lista):
    mostrar_info("Mascotas enfermas: ")
    encontrado = False
    for mascota in lista:
        if mascota["Salud"] == "Enfermo":
            mostrar_info(f"""
                 Nombre: {mascota["Nombre"]}
                 Especie: {mascota["Especie"]}
                 Edad: {mascota["Edad"]}
                 Peso: {mascota["Peso"]}
    ------------------------------------
                         Datos del dueño

                Nombre:{mascota["Dueño"]["Nombre"]}
                Dni:{mascota["Dueño"]["Dni"]}
                Telefono:{mascota["Dueño"]["Telefono"]}
                Dirección:{mascota["Dueño"]["Direccion"]}
        -------------------------------------------------
    """)
            encontrado = True

    if encontrado == False:
        mostrar_info("No se encontraron mascotas enfermas")


def actualizarSalud(lista):
    nombreABuscar = pedir_entrada("Nombre de mascota a buscar")
    encontrado = False
    for mascota in lista:
        if mascota["Nombre"] == nombreABuscar:
            nuevoEstado = pedir_entrada(f"""
                     Nombre: {mascota["Nombre"]}
                     Especie: {mascota["Especie"]}
                     Salud:{mascota["Salud"]}
        ----------------------
        Ingrese el nuevo estado de salud
        """)
            mascota["Salud"] = nuevoEstado
            encontrado = True
            break

    if encontrado == False:
        mostrar_info("No se encontró a la mascota")


def main():
    listaMascotas = []
    while True:
        try:
            opcion = int(pedir_entrada("""Seleccione una opción: 
                                     1: Anotar mascota
                                     2: Mostrar mascotas
                                     3: Buscar una mascota por su nombre
                                     4: Mostrar mascotas que pertenecen a un dueño
                                     5: Mostrar mascotas que se encuentran en estado "Enfermo",
                                     6: Actualizar el estado de salud de una mascota 
                                     7: Salir

                                     """
                                       ))
            if opcion:
                if opcion == 1:
                    cargarMascotas(listaMascotas)
                elif opcion == 2:
                    if len(listaMascotas) != 0:
                        mostrarMascotas(listaMascotas)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 3:
                    if len(listaMascotas) != 0:
                        buscarMascotaPorNombre(listaMascotas)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 4:
                    if len(listaMascotas) != 0:
                        mostrarMascotasDeTalDueño(listaMascotas)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 5:
                    if len(listaMascotas) != 0:
                        mostrarMascotasEnfermas(listaMascotas)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 6:
                    if len(listaMascotas) != 0:
                        actualizarSalud(listaMascotas)
                    else:
                        mostrar_error("La lista está vacía")

                elif opcion == 7:
                    mostrar_info("Saliendo.....")
                    break
            else:
                mostrar_info("Opción incorrecta")
        except ValueError:
            mostrar_error("Ingrese un número válido")


main()
```

```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import re


def pedir_entrada(prompt):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el cuadro de diálogo de entrada
    entrada = simpledialog.askstring(title="Entrada", prompt=prompt)

    # Cerrar la ventana principal
    root.destroy()

    return entrada


def mostrar_error(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de error
    messagebox.showerror("Error", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def mostrar_info(mensaje):
    # Crear la ventana principal (oculta)
    root = tk.Tk()
    root.withdraw()  # Ocultar la ventana principal

    # Mostrar el mensaje de información
    messagebox.showinfo("Información", mensaje)

    # Cerrar la ventana principal
    root.destroy()


def validarNum(i):
    while True:
        try:
            num = int(pedir_entrada(f"Número: {i+1}"))
            if 0 <= num:
                break
            else:
                mostrar_error(
                    "El número debe estar entre 0 y 20")
        except (ValueError):
            mostrar_error("No es un número entero")

    return num


def ingresarNum(lista):
    mostrar_info("Ingrese 5 números")
    for i in range(5):

        num = validarNum(i)
        lista.append(num)


def mostrarLista(lista):
    datos = ""
    for i in range(len(lista)):
        datos += f"{lista[i]}   "

    mostrar_info(f"""Lista de números
    {datos}
                 """)


def obtenerTresNumMayores(lista):
    listaOrdenada = lista.copy()
    listaOrdenada.sort()
    tresNumMayores = listaOrdenada[-4:]
    return tresNumMayores


def calcularSumMax(lista):
    suma = 0
    tresNumMayores = obtenerTresNumMayores(lista)
    for num in tresNumMayores:
        suma += num

    return suma


def obtenerTresNumMenores(lista):
    listaOrdenada = lista.copy()
    listaOrdenada.sort(reverse=True)
    tresNumMenores = listaOrdenada[-4:]
    return tresNumMenores


def calcularSumaMin(lista):
    suma = 0
    tresNumMenores = obtenerTresNumMenores(lista)
    for num in tresNumMenores:
        suma += num

    return suma


def main():
    lista = []
    ingresarNum(lista)
    mostrarLista(lista)
    sumaMayor = calcularSumMax(lista)
    sumaMenor = calcularSumaMin(lista)
    mostrar_info(f"""Mayor y menor suma que se puede obetener de la lista
    {sumaMayor}  {sumaMenor}
                 """)


main()
```
